```python
# OPIS: 
# Dane pobrałam z https://www.kaggle.com/datasets/dansbecker/melbourne-housing-snapshot
# Dane dotyczą nieruchomości w mieście Melbourne. Mamy do dyspozyjcji poniższe atrybuty:
# - ilość pokoji
# - cena za nieruchomośc w dolarach
# - metodę oznaczająca czy nieruchomość została sprzedana czy nie i w jaki sposób. Nieruchomości w Austalii są sprzedawane
# na aukcjach. Klienci licytują cenę, która jeśli nie osiągnie minimum ustalonego przez sprzedającego, powoduje, że
# nieruchomość nie zostaje sprzedana.
# ['S' - sprzedane 'SP' - sprzedane przed aukcją 'PI' - niesprzedane w akucji, cena niezadowalająca 'VB' 
# cena na auckcji nie osiągnęła oczekiwanego minimum przez sprzedającego i nieruchomość nie została sprzedana,
# 'SA' - sprzedane na aukcji]
# - rodzaj: br - mieszkanie, h- dom, u - bliźniak, t - dom w mieście, o - inne, dev sit - stan developerski
# - sprzedawcę
# - datę sprzedaży
# - odległość od ścisłego centrum
# - region
# - liczba nieruchomości w danej dzielnicy
# - liczba sypialni
# - liczba łazienek
# - liczba miejsc na samochód
# - wielkość gruntu
# - wielkość budynku
# - rada zarządzająca w danym miejscu
#
# Celem tego projektu jest zbudowanie modelu, który będzie w stanie określić najbardziej prawdopodobną cenę nieruchomości
# sprzedanej w 2016 lub 2017 roku
```

```python
# Jako, że nieruchomości były sprzedawane na przestrzeni dwóch lat to dla nieruchomości sprzedanych w roku 2016 
# przypisze 0 w kolumnie ['year_sold'] i 1 dla nieruchomości sprzedanych w 2017 roku. Ceny nieruchomości rosły w 
# tamtym czasie, dlatego ta kolumna ma dla modelu znaczenie - podobne nieruchomości sprzedane w 2017 powinny być
# droższe niż te sprzedane w 2016
```

```python
df['year_sold']= 0  # Inicjalizacja nowej kolumny wartościami 0
df.loc[df['Date'].dt.year == 2017, 'year_sold'] = 1  
```

```python
df.head()
```

```python
# Jako, że w tym projekcie chcę zbudowac model, kótry będzie w stanie przewidzieć realną cenę sprzedaży mieszkania, 
# usunę dane dla tych wierszy które mają w kolumnie 'method' PI i VB - jako, że te nieruchmości nie zostały sprzedane,
# więc cena była nierynkowa
```

```python
df2=df
df2=df2[df2['Method'] != 'PI']
df2=df2[df2['Method'] != 'VB']

df2.info()
```

```python
#Sprawdzam czy mam brakujące wartości
print("NaN values per column:\n", df2.isna().sum())
```

```python
# Dla niektórych rzędów brakuje informacji o miejscu dla samachodu przy nieruchomości. W tym przypadku jest to nieduży
# procent brakujących wartości do wszyskich rekordów, dlatego zamienię brakujące wartości na 0. Decyduję się na to rozwiązanie,
# bo skoro nie ma informacji na ten temat to możemy zakładać, że takich miejsc przy danej nieruchomości nie ma.
#
# Jeśli chodzi o Building Area to brakujących wartości jest na tyle dużo, że nie można ich usnąć, w takim razie zamierzam
# nie brać tego atrybutu pod uwagę w moim modelu. Istotniejsza informacją wydaje się liczba pokoji niż wielkość budynku.
#
#Council Area - ten atrybut też pomijam w dalszej analizie. Mamy dużo innch atrybutów opisujących lokalizację nieruchmości
#
#Year Build ten atrybut wydaje mi się istotny aczkolwiek ma również dużo brakujących wartości. Zastąpię brakujące wartości
# wartością średnią.

df2['Car']=df2['Car'].fillna(0)


df2['YearBuilt']=pd.to_datetime(df2['YearBuilt'])
df2['YearBuilt'] = df2['YearBuilt'].dt.year
df2['YearBuilt']=df2['YearBuilt'].fillna(df['YearBuilt'].mean())
```

```python
df2.head(10)
```

```python
print("NaN values per column:\n", df2.isna().sum())
```

```python
# Jako że mamy dużo atrybutów opisujących lokazlizację (address, suburb, distance from city center, council area,
# lattitude, longtitude, region name) zdecydowałam się wybrać tylko latitude i longtitude - najlepiej sa one w stanie oddać
# czy położenie nieruchomości jest atrakcyjne. Melbourn to duże miasto, którego centrum jest atrakcyjną lokalizacją, ale 
# też duża część miasta rozciąga się nad oceanem z dostępem do pięknych plaż i te dzielnice są równie atrakcyjne więc
# nieruchomości w różnych częściach miasta mogą być dużo warte.
#
# Wybór ten niestety powoduje, że zastosowanie regresji liniowej w tym przypadku nie jest możliwe, gdyż wartośc
# nieruchomości ze względu na jej położenie nie rośnie liniowo.
#
# Do dalszej analizy użyje następujących atrybutów: rooms, year_sold, type, date, bedroom2, bathroom, car, landsize, 
# yearbuilt, latitude i longtitude
#
y=df2['Price']

atrybuty=['Rooms', 'year_sold', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'YearBuilt', 'Lattitude', 'Longtitude']
X=df2[atrybuty]
```

# BUDOWANIE MODELU

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeRegressor(random_state=1)
model.fit(X_train, y_train)
```

```python
X_train
```

```python
y_train
```

# SPRAWDZENIE MODELU

```python
# Sprawdzam i liczę mean absolute error
from sklearn import metrics
import numpy as np

val_predictions = model.predict(X_test)
val_mae = mean_absolute_error(val_predictions, y_test)
print("MAE max_leaf_nodes: {:,.0f}".format(val_mae))

# Uzywam najlepszej wartości dla max_leaf_nodes
model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
model.fit(X_train, y_train)
val_predictions = model.predict(X_test)
val_mae = mean_absolute_error(val_predictions, y_test)
print("MAE dla najlepszej wartoścoi max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
# Teraz przetestuję Random Forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)

# Liczę mean absolute error
rf_val_predictions = rf_model.predict(X_test)
rf_val_mae = mean_absolute_error(rf_val_predictions, y_test)

print("MAE dla modelu Random Forest jest: {}".format(rf_val_mae))

#Za pomocą Random Forest uzyskuje lepsze wyniki, MAE jest niższe niż to uzyskane przez drzewo decyzyjne
```

```python
#Szukam najlepszych parametrów do modelu random forest:
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_regression
# Definicja siatki parametrów do przetestowania
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# Inicjalizacja obiektu GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring='neg_mean_absolute_error', cv=5)

# Uruchomienie procedury Grid Search
grid_search.fit(X, y)

# Wyświetlenie najlepszych parametrów i wyniku
print("Najlepsze parametry:", grid_search.best_params_)
print("Najlepszy wynik:", grid_search.best_score_)
```

```python
# Random Forest z najlepszymi parametrami: (nie wpisuje max depth jako None - z założenia mamy None)
rf_model = RandomForestRegressor(n_estimators=300, min_samples_split=2, random_state=1)
rf_model.fit(X_train, y_train)

# Liczę mean absolute error
rf_val_predictions = rf_model.predict(X_test)
rf_val_mae = mean_absolute_error(rf_val_predictions, y_test)

print("MAE dla modelu Random Forest jest: {}".format(rf_val_mae))
```

```python
# Teraz zrobię to samo ale z cross validation zamiast train test split
from sklearn.model_selection import cross_val_score
from sklearn import tree

rf_model = RandomForestRegressor(n_estimators=300, min_samples_split=5, random_state=1)

result = cross_val_score(rf_model, X, y, cv=5, scoring='neg_mean_absolute_error')

print(result)
```
